````python
# Import necessary libraries
import os
from IPython.display import display, Markdown, HTML
from dotenv import load_dotenv
from openai import OpenAI
from anthropic import Anthropic

# Load environment variables
load_dotenv()
print("Attempting to load API keys from .env file...")

# Load Keys
openai_api_key = os.getenv("OPENAI_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")
openrouter_api_key = os.getenv("OPENROUTER_API_KEY")  # ← This was missing!

# OpenRouter Client (used for ALL providers via OpenRouter)
openrouter_client = OpenAI(
    api_key=openrouter_api_key,        # ← Was wrongly using openai_api_key
    base_url="https://openrouter.ai/api/v1",
)
print(f"OpenRouter Client configured (Key starts with: {openrouter_api_key[:8]}...).")

# Direct OpenAI Client (bypasses OpenRouter, hits OpenAI directly)
openai_client = OpenAI(api_key=openai_api_key)
print(f"OpenAI Client configured (Key starts with: {openai_api_key[:5]}...).")
````

````python
# Helper function to display markdown nicely ---
def print_markdown(text):
    """Displays text as Markdown in Jupyter."""
    display(Markdown(text))
````

````python
def display_html_code(provider_name, html_content):
    """Displays generated HTML code block nicely."""
    print_markdown(f"### Generated HTML from {provider_name}:")
    # Display as a formatted code block
    display(Markdown(f"```html\n{html_content}\n```"))
````

````python
# Define the core prompt for the LLMs
# We explicitly ask for HTML code and specify the file name 'index.html'
html_prompt = f"""
Create ONLY the full HTML, starting with <!DOCTYPE html> and ending with </html>.

Design a modern, visually polished streaming platform landing page inspired by Netflix’s layout and visual style (do NOT copy logos, branding, or copyrighted assets).

Requirements:

A sleek, dark-themed header with navigation links (Home, TV Shows, Movies, New & Popular, My List) and a prominent Sign In button

A cinematic hero section with a large background-style layout, headline, short description, and primary call-to-action button

A content showcase section with multiple horizontal “carousel-style” rows of cards representing movies/shows

A features/value section emphasizing benefits like unlimited streaming, multi-device access, personalized recommendations, and offline viewing

A “How It Works” or “Why Choose Us” section highlighting simplicity and convenience

A testimonials or social-proof section with fictional user quotes

A pricing/subscription section with at least two plans

A professional footer with structured navigation links and social media icons

Styling & behavior:

Use inline CSS only (no external files)

Apply a modern dark color palette with accent highlights

Include responsive design for desktop and mobile

Add subtle animations, hover effects, and smooth transitions

Use clean typography, generous spacing, and a premium UI feel

Content focus:

Emphasize streaming convenience, personalization, and entertainment value

Use persuasive, conversion-oriented marketing copy

Highlight pain points (limited choices, ads, inconvenience) and solutions (on-demand access, recommendations, seamless experience)

Do not include explanations. Output ONLY the raw HTML code.
"""

print_markdown("**Core Prompt defined for the LLMs:**")
print_markdown(f"> {html_prompt}")  # Print the start of the prompt to verify
````

````python
openai_html_output = "<!-- OpenAI generation not run or failed -->"

print_markdown("## Calling OpenAI API...")
try:
    response = openrouter_client.chat.completions.create(  # ← openrouter_client
        model="openai/gpt-4o-mini",  # ← correct for OpenRouter
        messages=[
            {"role": "user", "content": html_prompt}
        ],
        temperature=0.5,
    )
    openai_html_output = response.choices[0].message.content

    # Strip markdown fences if present
    if openai_html_output.strip().startswith("```html"):
        lines = openai_html_output.strip().splitlines()
        openai_html_output = "\n".join(lines[1:-1]).strip()
    else:
        openai_html_output = openai_html_output.strip()

    display_html_code("OpenAI (gpt-4o-mini)", openai_html_output)

    file_path = "openai_landing_page.html"
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(openai_html_output)
    print_markdown(f"Successfully saved OpenAI output to `{file_path}`")

except Exception as e:
    print_markdown(f"Error calling OpenAI API: {e}")
    openai_html_output = f"<!-- Error calling OpenAI API: {e} -->"
````

````python
gemini_html_output = "<!-- Gemini generation not run or failed -->"

print_markdown("## Calling Google Gemini API...")
try:
    response = openrouter_client.chat.completions.create(  # ← changed
        model="google/gemini-2.5-flash-lite",             # ← cheap model
        max_tokens= 50000,             
        messages=[
            {"role": "user", "content": html_prompt}
        ],
    )

    raw_output = response.choices[0].message.content       # ← changed from response.text

    if raw_output.strip().startswith("```html"):
        lines = raw_output.strip().splitlines()
        gemini_html_output = "\n".join(lines[1:-1]).strip()
    else:
        gemini_html_output = raw_output.strip()

    display_html_code("Google Gemini (gemini-2.5-flash-lite)", gemini_html_output)

    file_path = "gemini_landing_page.html"
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(gemini_html_output)
    print_markdown(f"Successfully saved Gemini output to `{file_path}`")

except Exception as e:
    print_markdown(f"Error calling Google Gemini API: {e}")
    gemini_html_output = f"<!-- Error calling Google Gemini API: {e} -->"
````

````python
# Generate HTML using Anthropic Claude

claude_html_output = "<!-- Claude generation not run or failed -->"  # Default message

print_markdown("## Calling Anthropic Claude API...")

try:
    response = openrouter_client.chat.completions.create(
        model = "anthropic/claude-3-5-haiku",
        max_tokens = 4000,  # Set a max limit for the generated output
        # System prompt can sometimes help guide Claude's persona/role
        # system="You are a front-end web developer generating HTML code.",
        messages = [{"role": "user", "content": html_prompt}],
    )

    # Extract the text content from the response object
    raw_output = response.choices[0].message.content 
    if raw_output.strip().startswith("```html"):
        lines = raw_output.strip().splitlines()
        claude_html_output = "\n".join(lines[1:-1]).strip()
    else:
        claude_html_output = raw_output.strip()

    # Display the generated HTML code
    display_html_code(f"Anthropic Claude (claude-3-5-haiku)", claude_html_output)

    # --- Save the output to a file ---
    file_path = "claude_landing_page.html"

    with open(file_path, "w", encoding="utf-8") as f:
        f.write(claude_html_output)
    print_markdown(f"Successfully saved Claude output to `{file_path}`")

except Exception as e:
    print_markdown(f"Error calling Anthropic Claude API: {e}")
    claude_html_output = f"<!-- Error calling Anthropic Claude API: {e} -->"


````
